## Load Data

```python
train_df = pd.read_csv("/kaggle/input/Train.csv")
train_df.head(2)
```

```python
test_df = pd.read_csv("/kaggle/input/Test.csv")
test_df.head(2)
```

## Clean NaN values

```python
train_df.info()
```

```python
new_train_df = train_df.drop(['Attrition_rate', 'Employee_ID'], axis=1)
new_train_df.fillna(method='ffill', inplace=True)
new_train_df.info()
```

```python
test_df.info()
```

```python
new_test_df = test_df.drop('Employee_ID', axis=1)
new_test_df.fillna(method='ffill', inplace=True)
new_test_df.info()
```

## Dummy Encoding

```python
new_train_df = pd.get_dummies(new_train_df, drop_first=True)
new_train_df
```

```python
new_test_df = pd.get_dummies(new_test_df, drop_first=True)
new_test_df
```

## Model

```python
from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score
```

```python
print("XGBoost : ", cross_val_score(XGBRegressor(), new_train_df, train_df['Attrition_rate'], cv=10, scoring='neg_mean_squared_error').mean())
```

```python
model = XGBRegressor(n_estimators=500, learning_rate=0.05)
model.fit(new_train_df, train_df['Attrition_rate'], verbose=False)
```

```python
prediction = model.predict(new_test_df)

output = pd.DataFrame({'Employee_ID': test_df.Employee_ID, 'Attrition_rate': prediction})
output.to_csv('my_submission.csv', index=False)
```
