```python
import matplotlib.pyplot as plt
import cv2
```

```python
# Load data
train = pd.read_csv("/kaggle/input/identify-the-dance-form/train.csv")
test = pd.read_csv("/kaggle/input/identify-the-dance-form/test.csv")

train.head(5)
```

```python
# Image directories
train_dir = "/kaggle/input/identify-the-dance-form/train/"
test_dir = "/kaggle/input/identify-the-dance-form/test/"

train_files = os.listdir(train_dir)
test_files = os.listdir(test_dir)
```

### Train Set

```python
forms = ['manipuri','bharatanatyam','odissi','kathakali','kathak','sattriya','kuchipudi','mohiniyattam']
```

```python
# Load all train images in size 200 X 200

train_images = []
train_labels_temp = []
train_labels = []

for img in train_files:
    # Append Images after resizing
    train_images.append(cv2.resize(cv2.imread(train_dir + img), (200, 200), cv2.INTER_CUBIC)) # INTER_CUBIC as its efficient.
    # Append Labels
    if img in list(train['Image']):
        train_labels_temp.append(train.loc[train['Image'] == img, 'target'].values[0])
        

for i in train_labels_temp:
    train_labels.append(forms.index(i))
```

```python
# Visualize first 10 Images.

plt.figure(figsize = (15, 15))
for i in range(16):
    plt.subplot(5, 5, i+1)
    plt.imshow(train_images[i])
    plt.title(train_labels[i])
    plt.axis("off")
```

### Test Set

```python
test_images = []

for img in test_files:
    # Append Images after resizing
    test_images.append(cv2.resize(cv2.imread(test_dir + img), (200, 200), cv2.INTER_CUBIC)) # INTER_CUBIC as its efficient.
```

### Label Encode the target variable

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
train_labels = enc.fit_transform(train_labels)
```

### Convert training data to numpy array and reshape

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_images, train_labels, test_size=0.3, random_state=42)
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)

X_train = X_train/255
X_test = X_test/255
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Modelling

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense, Dropout, Flatten
```

```python
model = Sequential()

model.add(Flatten(input_shape=[200,200,3]))
model.add(Dense(100, activation = 'relu'))
model.add(Dense(50, activation = 'softmax'))

model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=20)
```

```python
test_images = np.array(test_images)
test_images = test_images/255
```

```python
pred = model.predict(test_images)
results = np.argmax(pred, axis=1)
output_sub = []
for i in results:
    output_sub.append(forms[i])
```

```python
output = pd.DataFrame({'Image': test.Image, 'target': output_sub})
output.to_csv('my_submission.csv', index=False)
```
